```python
# Zad 3
class TrieNode:
    def __init__(self, char):
        self.char = char
        self.next = {}
    
    def print(self):
        self._print(-1)

    def _print(self, indent):
        for i in range(indent):
            print("-", end="")
        print(self.char)
        for next_node in self.next.values():
            next_node._print(indent+1)

def create_trie(text):
    def find_head(node, suffix):
        for i in range(len(suffix)):
            next_node = node.next.get(suffix[i], None)
            if next_node is None: # next node not found
                # this is the head
                return i, node
            else: #next node found
                # go to next one
                node = next_node
        raise Exception("Error when searching for head")


    def add_suffix(node, suffix):
        for char in suffix:
            next_node = TrieNode(char)
            node.next[char] = next_node
            node = next_node
        

    trie = TrieNode("")
    for i in range(len(text)):
        suffix = text[i:]
        index, head = find_head(trie, suffix)
        add_suffix(head, suffix[index:])
    return trie
```

```python
# Zad 4

class SuffixTree:
    def __init__(self, text):
        self.text = text

        self.root = SuffixTreeNode(0, 0)
        for i in range(len(text)):
            self._add_suffix(self.root, i)

    def print(self):
        self.root.print()

    def _add_suffix(self, node, suffix_start):
        i = 0
        while True:
            if i == node.end - node.start: # reached end of node's text
                # Find next node
                next_node = node.next.get(self.text[suffix_start + i], None)
                if next_node is None: #Not found next node
                    # Create one
                    new_node = SuffixTreeNode(suffix_start + i, len(self.text))
                    node.next[self.text[suffix_start + i]] = new_node
                else: # Found next node
                    # Go to next one
                    self._add_suffix(next_node, suffix_start + i)
                return
            if self.text[node.start + i] != self.text[suffix_start + i]: #node and suffix text diverge
                # Split node into two nodes
                new_node = SuffixTreeNode(node.start + i, node.end)
                node.end = node.start + i
                
                new_node.next = node.next
                node.next = {self.text[node.start + i] : new_node}

                new_node = SuffixTreeNode(suffix_start + i, len(self.text))
                node.next[self.text[suffix_start + i]] = new_node
                return
            i += 1


class SuffixTreeNode:
    def __init__(self, start, end):
        self.next = {}
        self.start = start
        self.end = end

    def print(self):
        self._print(0)

    def _print(self, indent):
        for i in range(indent):
            print("-", end="")
        print(f"[{self.start}, {self.end}]")
        for char in self.next:
            self.next[char]._print(indent+1)
```
